```python
import pandas as pd
import plotnine as p9
import numpy as np
import scanpy as sc
```

```python
p9.options.dpi=300
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype']='none'
```

```python
OAK_March150k_gem_classification=pd.read_csv(OAK_150k_directoryname+'/outs/analysis/gem_classification.csv',header=0) 
```

```python
class_dict={"Multiplet":'Multiplet_obs'} 
OAK_March150k_gem_classification.replace({"call":class_dict},inplace=True) 
```

```python
OAK_March150k_gem_classification['call'].value_counts()
```

```python
OAK_March150k_gem_classification.to_csv('SupFig1d_left.csv',header=True,index=False)
```

```python
plot=p9.ggplot(data=OAK_March150k_gem_classification,
               mapping=p9.aes(x='GRCh38',y='mm10',color="call")
              )
plot+ p9.geom_point(size=0.1,alpha=0.03) \
+ p9.themes.theme_bw() \
+ p9.coord_fixed(ratio=1,xlim=(0,30000),ylim=(0,30000)) \
+ p9.theme(axis_text_x=p9.element_text(colour="black",angle = 45, hjust = 1)) \
+ p9.theme(axis_text_y=p9.element_text(colour="black")) \
+ p9.labels.ylab('Mouse UMI') \
+ p9.labels.xlab('Human UMI') \
+ p9.theme(text=p9.element_text(size=7),legend_position=(0.66,0.75),legend_title=p9.element_blank(),legend_direction='vertical',legend_key=p9.element_blank(),legend_key_size=5,legend_box_margin=0) \
+ p9.guides(color = p9.guide_legend(override_aes={'size' :1,'alpha':0.5 })) \
+ p9.scale_color_manual(values = {'GRCh38':'#E24B15','mm10':'#5972CB','Multiplet_obs':'#11f505'}) \
+ p9.theme(figure_size=(2,2))
```

```python
plot=p9.ggplot(data=OAK_March150k_gem_classification,
               mapping=p9.aes(x='GRCh38',y='mm10',color="call")
              )
plotsave=plot+ p9.geom_point(size=0.1,alpha=0.03) \
+ p9.themes.theme_bw() \
+ p9.coord_fixed(ratio=1,xlim=(0,30000),ylim=(0,30000)) \
+ p9.theme(axis_text_x=p9.element_text(colour="black",angle = 45, hjust = 1)) \
+ p9.theme(axis_text_y=p9.element_text(colour="black")) \
+ p9.labels.ylab('Mouse UMI') \
+ p9.labels.xlab('Human UMI') \
+ p9.theme(text=p9.element_text(size=7),legend_position=(0.66,0.75),legend_title=p9.element_blank(),legend_direction='vertical',legend_key=p9.element_blank(),legend_key_size=5,legend_box_margin=0) \
+ p9.guides(color = p9.guide_legend(override_aes={'size' :1,'alpha':0.5 })) \
+ p9.scale_color_manual(values = {'GRCh38':'#E24B15','mm10':'#5972CB','Multiplet_obs':'#11f505'}) \
+ p9.theme(figure_size=(2,2))
fig=plotsave.draw()
fig.set_size_inches(2, 2)
points = fig.axes[0].collections[0]
points.set_rasterized(True)
fig.savefig('OAK_150k_barnyard.svg',format='svg',dpi=300,bbox_inches='tight')  
```
