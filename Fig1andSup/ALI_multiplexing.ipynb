```python
plot=p9.ggplot(data=ALI_HTOpercent,
               mapping=p9.aes(x='Percentage.OAK',y='Percentage.Control',color='Hashtag')
              )
plot+ p9.geom_point(alpha=1,size=1,show_legend=True) \
+ p9.themes.theme_bw() \
+ p9.theme(axis_text_x=p9.element_text(angle = 45, hjust = 1)) \
+ p9.theme(text=p9.element_text(size=7)) \
+ p9.scales.scale_x_continuous(limits=[6,16.5],breaks=np.arange(6,16.5,2)) \
+ p9.scales.scale_y_continuous(limits=[6,16.5],breaks=np.arange(6,16.5,2)) \
+ p9.labels.xlab('OAK assignment (%)') \
+ p9.labels.ylab('Chromium assignment (%)') \
+ p9.geom_abline(intercept=0,slope=1,linetype='dashed',color='grey') \
+ p9.theme(figure_size=(2,2)) 
```

```python
plot=p9.ggplot(data=ALI_HTOpercent,
               mapping=p9.aes(x='Percentage.OAK',y='Percentage.Control',color='Hashtag')
              )
pl=plot+ p9.geom_point(alpha=1,size=1,show_legend=False) \
+ p9.themes.theme_bw() \
+ p9.theme(axis_text_x=p9.element_text(angle = 45, hjust = 1)) \
+ p9.theme(text=p9.element_text(size=7)) \
+ p9.scales.scale_x_continuous(limits=[6,16.5],breaks=np.arange(6,16.5,2)) \
+ p9.scales.scale_y_continuous(limits=[6,16.5],breaks=np.arange(6,16.5,2)) \
+ p9.labels.xlab('OAK assignment (%)') \
+ p9.labels.ylab('Chromium assignment (%)') \
+ p9.geom_abline(intercept=0,slope=1,linetype='dashed',color='grey') \
+ p9.theme(figure_size=(2,2)) 
fig=pl.draw()
fig.set_size_inches(2, 2)
fig.savefig('OAK_ALI_multiplex.svg',format='svg',dpi=300,bbox_inches='tight')  
```
