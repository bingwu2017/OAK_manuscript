```python
import pandas as pd
import numpy as np
import plotnine as p9
```

```python
p9.options.dpi=300
```

```python
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype']='none'
```

```python
def collion_num(N,D):
    return N-D+(D*pow((D-1)/D,N))
```

```python
collion_num(150000,100000*12)/150000
```

```python
collion_num(450000,100000*40)/(450000)
```

```python
collision_rate_150k=np.array([])
for i in range(1,61):
    collision_rate_150k=np.append(collision_rate_150k,collion_num(150000,100000*i)/150000)
```

```python
collision_rate_150k
```

```python
collision_rate_450k=np.array([])
for i in range(1,61):
    collision_rate_450k=np.append(collision_rate_450k,collion_num(450000,100000*i)/450000)
```

```python
collision_rate_450k
```

```python
df_collision_rate_150k=pd.DataFrame()
```

```python
df_collision_rate_150k['num_aliquots']=range(1,61)
```

```python
df_collision_rate_150k['collision_rate']=collision_rate_150k*100
```

```python
df_collision_rate_150k['loading']='150k'
```

```python
df_collision_rate_450k=pd.DataFrame()
```

```python
df_collision_rate_450k['num_aliquots']=range(1,61)
```

```python
df_collision_rate_450k['collision_rate']=collision_rate_450k*100
```

```python
df_collision_rate_450k['loading']='450k'
```

```python
df_collision_rate=pd.concat([df_collision_rate_150k,df_collision_rate_450k])
```

```python
pd.set_option('display.max_rows', 500)
```

```python
df_collision_rate
```

```python
colors={'150k':'#B5D251','450k':'#F7BB2C'}
```

```python
plot=p9.ggplot(data=df_collision_rate,
               mapping=p9.aes(x='num_aliquots',y='collision_rate',color='loading')
              )
plot+ p9.geom_point(size=0.6,alpha=0.6) \
+ p9.themes.theme_bw() \
+ p9.theme(axis_text_x=p9.element_text(rotation = 90)) \
+ p9.labels.xlab('Number of aliquots') \
+ p9.labels.ylab('Collision rate (%)') \
+ p9.scales.scale_y_continuous(breaks=np.arange(0,100,10)) \
+ p9.labs(color='') \
+ p9.theme(text=p9.element_text(size=7),legend_position=(0.7,0.7),legend_title=p9.element_blank(),legend_direction='vertical',legend_key=p9.element_blank(),legend_key_size=5) \
+ p9.scale_color_manual(values = colors) \
+ p9.theme(figure_size=(3,3)) # 3 inch
```

```python
plot=p9.ggplot(data=df_collision_rate,
               mapping=p9.aes(x='num_aliquots',y='collision_rate',color='loading')
              )
plotsave=plot+ p9.geom_point(size=0.6,alpha=0.6) \
+ p9.themes.theme_bw() \
+ p9.theme(axis_text_x=p9.element_text(rotation = 90)) \
+ p9.labels.xlab('Number of aliquots') \
+ p9.labels.ylab('Collision rate (%)') \
+ p9.scales.scale_y_continuous(breaks=np.arange(0,100,10)) \
+ p9.labs(color='') \
+ p9.theme(text=p9.element_text(size=7),legend_position=(0.7,0.7),legend_title=p9.element_blank(),legend_direction='vertical',legend_key=p9.element_blank(),legend_key_size=5) \
+ p9.scale_color_manual(values = colors) \
+ p9.theme(figure_size=(3,3)) # 3 inch
plotsave.save('CollisionEstimation.svg',height=3,width=3,format='svg',dpi=300) 
```
