```python
import pandas as pd
import plotnine as p9
import numpy as np
```

```python
p9.options.dpi=300
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype']='none'
```

```python
OAK_150k=pd.read_csv(data_dir+'barnyard_OAK150kali1/count_summary.csv',header=0)
```

```python
OAK_450k=pd.read_csv(data_dir+'barnyard_OAK450kali1/count_summary.csv',header=0)
```

```python
OAK_150k_toConcat=OAK_150k[['cell_bc','call','Read_count','n_genes_by_counts','total_counts','exp_group']]
```

```python
OAK_150k_toConcat.columns=['cell_bc','species','read_count','n_genes_by_counts','n_UMI_counts','exp_group']
```

```python
OAK_450k_toConcat=OAK_450k[['cell_bc','call','Read_count','n_genes_by_counts','total_counts','exp_group']]
```

```python
OAK_450k_toConcat.columns=['cell_bc','species','read_count','n_genes_by_counts','n_UMI_counts','exp_group']
```

```python
ExpGroup_Compound=pd.concat([OAK_150k_toConcat,OAK_450k_toConcat])
```

```python
di={'OAK_150kMar':'150000','OAK_450kMar':'450000'}
```

```python
ExpGroup_Compound=ExpGroup_Compound.replace({'exp_group':di})
```

```python
plot=p9.ggplot(data=ExpGroup_Compound[ExpGroup_Compound['species']=='GRCh38'],
               mapping=p9.aes(x='read_count',y='n_genes_by_counts',color='exp_group')
              )
plot+ p9.geom_point(alpha=0.3,size=0.1,show_legend=True) \
+ p9.themes.theme_bw() \
+ p9.theme(axis_text_x=p9.element_text(angle = 45, hjust = 1)) \
+ p9.scale_color_manual(values = {'150000':'#B5D251','450000':'#F7BB2C'}) \
+ p9.theme(text=p9.element_text(size=7),legend_title=p9.element_blank(),legend_position=(.3, 0.75),legend_direction='vertical',legend_key=p9.element_blank(),legend_box_margin=0) \
+ p9.guides(color = p9.guide_legend(override_aes={'size' :1,'alpha':0.5 })) \
+ p9.ylim(0,8500) \
+ p9.xlim(0,80000) \
+ p9.labels.ylab('Number of genes') \
+ p9.labels.xlab('Read count') \
+ p9.theme(figure_size=(2,2)) 

```

```python
plot=p9.ggplot(data=ExpGroup_Compound[ExpGroup_Compound['species']=='GRCh38'],
               mapping=p9.aes(x='read_count',y='n_genes_by_counts',color='exp_group')
              )
plotsave=plot+ p9.geom_point(alpha=0.3,size=0.1,show_legend=True) \
+ p9.themes.theme_bw() \
+ p9.theme(axis_text_x=p9.element_text(angle = 45, hjust = 1)) \
+ p9.scale_color_manual(values = {'150000':'#B5D251','450000':'#F7BB2C'}) \
+ p9.theme(text=p9.element_text(size=7),legend_title=p9.element_blank(),legend_position=(.3, 0.75),legend_direction='vertical',legend_key=p9.element_blank(),legend_box_margin=0) \
+ p9.guides(color = p9.guide_legend(override_aes={'size' :1,'alpha':0.5 })) \
+ p9.ylim(0,8500) \
+ p9.xlim(0,80000) \
+ p9.labels.ylab('Number of genes') \
+ p9.labels.xlab('Read count') \
+ p9.theme(figure_size=(2,2)) 
fig=plotsave.draw()
fig.set_size_inches(2, 2)
points = fig.axes[0].collections[0]
points.set_rasterized(True)
fig.savefig('OAK_human_loadingCurve.svg',format='svg',dpi=300,bbox_inches='tight')  
```
