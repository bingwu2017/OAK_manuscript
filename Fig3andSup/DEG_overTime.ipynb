```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import matplotlib.patches as mplpatches
```

```python
plt.rcParams['svg.fonttype']='none' 
```

```python
sc.settings.verbosity = 0             # verbosity: errors (0), warnings (1), info (2), hints (3)
```

```python
adata=sc.read_h5ad(project_dir+'/Anndata/T0T3T5Clone1_combined.h5ad')
```

```python
fc_lineage=pd.read_csv(D0data_dir+'/D0toD20_lineage_change.csv',header=0)
```

```python
adata.obs=adata.obs.reset_index().merge(fc_lineage,how="left",left_on="Lineage_assigned",right_on="LineageID").set_index('index')
```

```python
DayMap={'D0':'Day0','D10':'Day10','D20':'Day20','Resist_clone1':'Day90'}
```

```python
adata.obs['Day']=adata.obs['sample'].map(DayMap)
```

```python
sc.settings.set_figure_params(dpi=300, dpi_save=300, format='svg')
```

```python
fig,ax = plt.subplots(figsize=(0.9,1),dpi=300)
ax.set_ylim(0, 5)
ax.tick_params(axis='both', which='major', labelsize=7)
ax.yaxis.label.set_size(7)
sc.pl.violin(adata[adata.obs['D20overD0']=='Most_depleted',:], ['FN1'],
     groupby='Day',rotation=45,size=0.2,ax=ax,save='_FN1_MostDepleted')
```
